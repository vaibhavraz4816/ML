# Implement linear regression with multiple variables through gradient decent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Define the data
df = pd.read_csv('house_price.csv')
df.head() # To get first n rows from the dataset default value of n is 5

X = df.iloc[:,0:2]
y = df.iloc[:,2]

m = len(y)

X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
# Define the hyperparameters
learning_rate = 0.01
n_iterations = 1000

# Initialize the parameters
theta = np.zeros(X.shape[1])

# Define the cost function
def cost_function(X, y, theta):
    m = len(y)
    J = 1/(2*m)*np.sum((X.dot(theta) - y)**2)
    return J

# Gradient descent
J_history = []
for i in range(n_iterations):
    # Calculate the hypothesis
    h = X.dot(theta)
    
    # Calculate the gradient
    gradient = 1/len(y) * X.T.dot(h - y)
    
    # Update the parameters
    theta -= learning_rate * gradient
    
    # Calculate the cost function and add it to the history
    J_history.append(cost_function(X, y, theta))

# Print the final parameters
print("theta =", theta)

# Plot the cost function history
plt.plot(J_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
X = data['Hours'].values.reshape(-1,1)
y = data['Scores'].values.reshape(-1,1)

# Normalize the data
X = (X - np.mean(X)) / np.std(X)

# Add a column of ones to X for the intercept term
X = np.hstack((np.ones((X.shape[0],1)), X))

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define the cost function
def cost_function(X, y, theta):
    m = len(y)
    h = sigmoid(X.dot(theta))
    J = -1/m * (y.T.dot(np.log(h)) + (1-y).T.dot(np.log(1-h)))
    return J

# Gradient descent
n_iterations = 1000
alpha = 0.01
theta = np.zeros((X.shape[1],1))
J_history = []

for i in range(n_iterations):
    # Calculate the hypothesis and gradient
    h = sigmoid(X.dot(theta))
    gradient = 1/len(y) * X.T.dot(h - y)
    
    # Update the parameters
    theta -= alpha * gradient
    
    # Calculate the cost function
    J = cost_function(X, y, theta)
    J_history.append(J)
    
    # Print the cost every 100 iterations
    if i % 100 == 0:
        print("Iteration", i, "Cost =", J)
        
# Plot the cost vs iterations
plt.plot(range(n_iterations), J_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

# Predict the output for a new input
new_input = 7
new_input_normalized = (new_input - np.mean(X[:,1])) / np.std(X[:,1])
new_output = sigmoid(np.array([1, new_input_normalized]).dot(theta))
print("Predicted output for input", new_input, "is", new_output)
```

# Implement Logistic regression with multiple variables through gradient decent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
X = data[['Hours', 'Age']].values
y = data['Pass'].values.reshape(-1,1)

# Normalize the data
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Add a column of ones to X for the intercept term
X = np.hstack((np.ones((X.shape[0],1)), X))

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define the cost function
def cost_function(X, y, theta):
    m = len(y)
    h = sigmoid(X.dot(theta))
    J = -1/m * (y.T.dot(np.log(h)) + (1-y).T.dot(np.log(1-h)))
    return J

# Gradient descent
n_iterations = 1000
alpha = 0.01
theta = np.zeros((X.shape[1],1))
J_history = []

for i in range(n_iterations):
    # Calculate the hypothesis and gradient
    h = sigmoid(X.dot(theta))
    gradient = 1/len(y) * X.T.dot(h - y)
    
    # Update the parameters
    theta -= alpha * gradient
    
    # Calculate the cost function
    J = cost_function(X, y, theta)
    J_history.append(J)
    
    # Print the cost every 100 iterations
    if i % 100 == 0:
        print("Iteration", i, "Cost =", J)
        
# Plot the cost vs iterations
plt.plot(range(n_iterations), J_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

# Predict the output for a new input
new_input = np.array([7, 20])
new_input_normalized = (new_input - np.mean(X[:,1:], axis=0)) / np.std(X[:,1:], axis=0)
new_input_normalized = np.hstack((np.ones((1,1)), new_input_normalized))
new_output = sigmoid(new_input_normalized.dot(theta))
print("Predicted output for input", new_input, "is", new_output)
```

# Normal Method one Variable

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate some random data
np.random.seed(0)
X, y = make_classification(n_features=1, n_redundant=0, n_informative=1,
                             n_clusters_per_class=1, random_state=0)

# Create a LogisticRegression object
lr = LogisticRegression(solver='newton-cg', C=1e5)

# Fit the model to the data
lr.fit(X, y)

# Plot the decision boundary
plt.scatter(X[:, 0], y, c=y, cmap=plt.cm.Paired)
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
xx = np.linspace(x_min, x_max, 10)
Z = lr.predict_proba(xx.reshape(-1, 1))[:, 1]
plt.plot(xx, Z, color='blue', linewidth=3)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

# Two Variable Normal Method

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate some random data
np.random.seed(0)
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1, random_state=0)

# Create a LogisticRegression object
lr = LogisticRegression(solver='newton-cg', C=1e5)

# Fit the model to the data
lr.fit(X, y)

# Plot the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 10), np.linspace(y_min, y_max, 10))
Z = lr.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.5)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
