```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns

%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv("train.csv")
df =df.dropna()
df=df.reset_index(drop=True)
df.head()
```

```python
X = df.x
Y = df.y
```

```python
plt.figure()
g = plt.scatter(X,Y,marker='.')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original data')
plt.show()
```

```python
X_mean = np.mean(X)
Y_mean = np.mean(Y)
print((X_mean,Y_mean))
```

```python
tX = np.array([x - X_mean for x in X])
tY = np.array([y - Y_mean for y in Y])
```

```python
Sxy = sum(tX*tY)
Sx = sum(tX**2)
print((Sxy,Sx))
```

```python
Q1 = Sxy/Sx
Q0 = Y_mean - Q1*X_mean
print((Q0,Q1))
```

```python
H = Q0 + Q1*X
m = len(X)
J = 0
for h,y in zip(H,Y):
    J += (h - y) ** 2
    
J /= 2*m
print(J)
```

```python
fig=figure(figsize=(20,6),dpi=90)

a=fig.add_subplot(1,2,1)
a.set_title("Original Data")
a=plt.scatter(X,Y,marker=".")
plt.xlabel('X')
plt.ylabel('Y')

b=fig.add_subplot(1,2,2)
b.set_title("Predicted Data")
b=plt.scatter(X,H,marker=".",color='g')
plt.xlabel('X')
plt.ylabel('Y_pred')
plt.show()
```

```python
mse = 2*J
mse
```

```python
rmse = np.sqrt(mse)
rmse
```
